```python
import tensorflow as tf 
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
    #print(sentence)
    #print(tokenizer.texts_to_sequences([sentence])[0]) #converting the sentences into numbers
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    
    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
    
```

```python
input_sequences #creating a input sequence...we havent yet converted it into input and output format as of yet
```

```python
#We need to make the shape same using ZERO PADDING
```

```python
[len(x) for x in input_sequences] #length of each input sequences
```

```python
max_len = max([len(x) for x in input_sequences ])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences , maxlen=max_len, padding ='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test = train_test_split(X,test_size = 0.2,random_state=42)
```

```python
X.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
len(tokenizer.word_index)
```

```python
y =to_categorical(y,num_classes=901) #word indexing starts from 1
```

```python
y.shape
```

```python
y_train,  y_test = train_test_split(y,test_size=0.2,random_state=42)
```

```python
y_train.shape
y_test.shape
```

```python
y_train
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Embedding,LSTM,Dense,Bidirectional
```

```python
model = Sequential()
model.add(Embedding(901,100,input_length=15))
model.add(LSTM(150,return_sequences=True))
model.add(LSTM(150))
model.add(Dense(901,activation='softmax'))
```

```python
from tensorflow import keras
from keras.utils.vis_utils import plot_model

keras.utils.plot_model(model, to_file='model.png', show_layer_names=True)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import TensorBoard

checkpoint = ModelCheckpoint("nextword1.h5", monitor='loss', verbose=1,
    save_best_only=True, mode='auto')

reduce = ReduceLROnPlateau(monitor='loss', factor=0.2, patience=3, min_lr=0.0001, verbose = 1)

logdir='logsnextword1'
tensorboard_Visualization = TensorBoard(log_dir=logdir)
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.001),metrics=['accuracy'])
```

```python
model.summary()
```

```python
        history = model.fit(X_train,y_train,epochs=200,batch_size=128,callbacks=[checkpoint, reduce, tensorboard_Visualization],validation_data=(X_test,y_test))
```

```python
import matplotlib.pyplot as plt


def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.xlabel("Epochs")
    plt.ylabel(string)
    plt.show()
```
